```python
import numpy as np
import pandas as pd

# 이미 선언한 system_prompt, user_prompt_template
system_prompt = """당신은 한국의 문화와 관련된 문제를 전문적으로 풀이해주는 문제 해설가입니다.  
사용자가 입력한 문제에 대해 단계별로 차근차근(step by step) 설명하여 **문제 해설**과 **정답**을 제시하세요.  

답변 형식은 반드시 다음과 같이 작성하세요:  
문제 해설: (문제에 대한 배경, 개념 설명, 선택지 또는 정답 후보 분석 등 단계별로 논리적인 추론 과정을 포함)  
정답: (정답만 간결하게 작성)

※ 정답 작성 형식 안내  
- 선다형 문제일 경우: **정답 번호**만 작성하세요.  
- 단답형 문제일 경우:  
    - 정답이 1개인 경우: 괄호, 한자 없이 **한글** 또는 **영어**를 사용하세요. (예: 사과 또는 apple)  
    - 정답이 여러 개인 경우: 쉼표(,)로 구분해 나열하세요. (예: 사과, 배)  
    - 순서 배열 문제인 경우: '-'로 구분해 정확한 순서를 유지해 나열하세요. (예: ㄱ-ㄴ-ㄷ-ㄹ)"""
user_prompt_tpl = """아래 문제를 단계별로 자세히 해설해주고, 마지막에 정답을 작성해줘.  

키워드: {topic_keyword}  
문제 유형: {question_type}  
문제: {question}"""

# 각 행(row)을 받아 원하는 배열(array)을 반환하는 함수
def build_prompt(row):
    user_content = user_prompt_tpl.format(
        topic_keyword=row.get("topic_keyword", ""),
        question_type=row.get("question_type", ""),
        question=row.get("question", ""),
    )
    return np.array([                      # ← 예시 형식 그대로
        {"content": system_prompt, "role": "system"},
        {"content": user_content, "role": "user"}
    ], dtype=object)

# ★ DataFrame에 prompt 컬럼 추가
train_df["prompt"] = train_df.apply(build_prompt, axis=1)
filtered_df["prompt"] = filtered_df.apply(build_prompt, axis=1)
```

```python
train_df['reward_model'] = train_df['answer'].apply(lambda answer : np.array([                      # ← 예시 형식 그대로
        {'ground_truth': answer, 'style': 'rule'}
    ], dtype=object))

filtered_df['reward_model'] = filtered_df['answer'].apply(lambda answer : np.array([                      # ← 예시 형식 그대로
        {'ground_truth': answer, 'style': 'rule'}
    ], dtype=object))
train_df
```

```python
train_df['extra_info'] = train_df['question_type'].apply(lambda question_type : {'question_type':question_type})
filtered_df['extra_info'] = filtered_df['question_type'].apply(lambda question_type : {'question_type':question_type})
```

```python
train_df = train_df.rename(columns={'question_type':'data_source', 'topic_keyword':'ability'})
filtered_df = filtered_df.rename(columns={'question_type':'data_source', 'topic_keyword':'ability'})
train_df
```
